### f(x) = x^3 - 50x
### g(x) = -x^4 + 88x^2 - 241

```python
from sympy import *
from decimal import Decimal
```

## Найти координаты точек пересечения

```python

x = Symbol('x')

f = collect(x**3-50*x, x)
g = collect(-x**4 + 88*x**2 - 241, x)
h = collect(f - g, x)

intersec_x = solve(h, x)
print(f'Всего точек пересечения: {len(intersec_x)}')
```

```python
intersec_y = []
for i in intersec_x:
    intersec_y.append(f.subs(x, i))

print('x1 = ')
intersec_x[0]
```

```python
print('y1 = ')
intersec_y[0]
```

```python
print('x2 = ')
intersec_x[1]
```
